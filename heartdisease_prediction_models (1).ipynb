```python
# IMPORTING THE DEPENDIES
import numpy as np
import pandas as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# data processing and data collection
import pandas as pd
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

1 = defective heart

```python
# splitting the feature and the target
x = heart_data.drop(columns='target',axis=1)
y = heart_data['target']
print(x)
```

```python
print(y)
```

```python
# split the data in testing data and the training data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x.shape,x_train.shape,x.shape,x_test.shape)
```

```python
# model logistic regression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#EVALUATION OF THE MODEL
#ACCURACY OF THE TRAINING DATA
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("accuracy of trainig data:",training_data_accuracy)
```

```python
# ACCURACY OF THE TESTING DATA
x_test_prediction = model.predict(x_test)
testing_data_accuracy =accuracy_score(x_test_prediction,y_test)
print("accuracy of testing data:",testing_data_accuracy)
```
